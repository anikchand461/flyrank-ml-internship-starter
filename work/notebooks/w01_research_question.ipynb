## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
# Research question:
# Which content pages should be prioritized for human review based on observable
# search, content, and engagement signals?
#
# Decision:
# Which pages should a content team review first when review capacity is limited?
#
# Action:
# A content reviewer examines the highest-ranked pages first and decides whether
# each page should be refreshed, expanded, protected, pruned, or monitored.
#
# Cost of a wrong call:
# A false positive wastes limited reviewer time on a page that may not need
# attention. A false negative may cause the team to miss a page that genuinely
# deserves review.
```
