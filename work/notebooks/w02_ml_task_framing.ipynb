## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
# My chosen lane is Refresh / Content Opportunity Scoring.

# I would frame this as a ranking and scoring problem.

# The goal is to assign an opportunity score to each content page and rank
# pages based on how urgently they should be reviewed for a possible content
# refresh.

# The output of the system would help the content team decide which pages
# should be reviewed first.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
# The ideal target would be the expected improvement in search and engagement
# performance after refreshing a page.

# However, historical refresh outcome data may not be available.

# Therefore, I would use a proxy target called `refresh_opportunity_score`.

# This score could be estimated using observable signals such as:

# - impressions
# - traffic trend
# - engagement signals
# - content age
# - search visibility
# - recent performance decline

# A higher score indicates that a page is a stronger candidate for human review
# and possible refresh.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# ## 3. Success metric

# Since this is primarily a ranking problem, I would use Precision@K as the
# main success metric.

# For example, Precision@50 would measure how many of the top 50 pages
# recommended by the system are actually judged by the content team to be
# valuable refresh opportunities.

# This metric is useful because the content team has limited review capacity
# and mainly cares about the quality of the highest-ranked recommendations.
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

pages = pd.DataFrame({
    "page_id": [101, 102, 103, 104, 105],
    "impressions": [15000, 8000, 22000, 3000, 18000],
    "traffic_change_pct": [-35, -10, -42, 5, -28],
    "engagement_rate": [0.31, 0.52, 0.25, 0.68, 0.38],
    "content_age_days": [720, 300, 900, 120, 600],
    "refresh_opportunity_score": [0.91, 0.45, 0.96, 0.12, 0.82]
})

pages
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# A fixed rule such as:

# "Refresh every page with more than 500 impressions and declining traffic"

# would be too simplistic.

# In the available data, 16,726 pages have 500+ impressions and 16,262 pages
# show downward movement. A simple threshold rule could therefore produce a
# very large review queue.

# Machine learning can combine multiple signals such as traffic decline,
# impressions, engagement, and content age.

# It can learn interactions between these signals and produce a continuous
# priority score.

# The ranked output allows the content team to review the highest-opportunity
# pages first instead of manually checking thousands of pages.
```
